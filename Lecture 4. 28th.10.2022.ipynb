```python
name_list= ['Lobster star',
             "Chicken caer",
             "Bang bang",
             "Ham and cheese",
             "Tuna and cucumber",
             "Egg sandwich"
           ]
#namelist(categorical data) is a list of names and pyplot has arraned them in a certain order
#lines are plotted with continous data.  data types/types of variables are binary, discrete,categorical,continous, numerical 
boston_price_list= [9.99,7.99,7.49,7,6.29,4.99]
london_price_list= [7.5,5,4.4,5, 3.75, 2.25]
```

```python
plt.plot(boston_price_list, name_list)
plt.xlabel("Price in USD");
```

```python
plt.plot(boston_price_list, name_list,marker= "o",linestyle="" ) # by assigning it a marker we are removing a line but still in the line shape, 
#so we have separated the diff categories
plt.xlabel("Price in USD")
```

```python
plt.plot(boston_price_list, name_list,marker= "o",linestyle="" ) # by assigning it a marker we are removing a line
#but still in the line shape, so we have separated the diff categories
plt.plot(london_price_list, name_list,marker= "s",linestyle="" );
plt.xlabel("Price in USD")
plt.title("Food prices in Boston and London")
```

```python
plt.plot(boston_price_list, name_list,marker= "o", linestyle="" ,color="C3" ) # by assigning it a marker we are removing a line
#but still in the line shape, so we have separated the diff categories
plt.plot(london_price_list, name_list,marker= "s",linestyle="", color="C0" );
plt.xlabel("Price in USD")
plt.title("Food prices in Boston and London");
```

```python
plt.plot(boston_price_list, name_list,marker= "o", linestyle="" ,color="C3" ) # by assigning it a marker we are removing a line
#but still in the line shape, so we have separated the diff categories
plt.plot(london_price_list, name_list,marker= "s",linestyle="", color="C0" );
plt.xlabel("Price in USD")
plt.title("Food prices in Boston and London");
plt.hlines(name_list,london_price_list, boston_price_list);
```

```python
plt.plot(london_price_list, name_list,marker= "s",linestyle="", color="C0", label="London price list" );
plt.plot(boston_price_list, name_list,marker= "o", linestyle="" ,color="C3" ,label="Boston price list") # by assigning it a marker we are removing a line
#but still in the line shape, so we have separated the diff categories
plt.xlabel("Price in USD")
plt.title("Food prices in Boston and London");
plt.hlines(name_list,london_price_list, boston_price_list);
plt.legend();
```
